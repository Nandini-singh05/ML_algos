```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Dell\Downloads\homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price', axis='columns')
new_df
```

```python
price=df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

#### 1)Predict price of a home with area = 3300 sqr ft

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

#### y = m*x + b (m->coefficient and b->intercept)

```python
3300 * 135.78767123 + 180616.43835616432
```

#### 2)Predict price of a home with area = 5000 sqr ft

```python
reg.predict([[5000]])
```

### Generate CSV file with list of home price predictions 

```python
area_df = pd.read_csv(r"C:\Users\Dell\Downloads\areas.csv")
area_df.head(3)
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['price'] = p
area_df
```

```python
area_df.to_csv('prediction.csv')
```

## Exercise 
Predict canada's per capita income in year 2020. 


```python
df2 = pd.read_csv(r'C:\Users\Dell\Downloads\canada_per_capita_income.csv')
df2
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.scatter(df2.year, df2[['per capita income (US$)']], color='blue', marker='*')
```

```python
new_df2 = df2.drop(df2[['per capita income (US$)']], axis='columns')
new_df2
```

```python
per_capita_income = df2[['per capita income (US$)']]
per_capita_income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df2, per_capita_income)
```

```python
predicted_income = reg.predict([[2020]])
```

```python
print(f'The predicted per capita income in 2020 is : {predicted_income}')
```
